```python
#백터화된 함수 사용하기
import pandas as pd

df = pd.DataFrame({"a": [10, 20, 30], 
                   "b": [20, 30, 40]})
print(df)
```

```python
#평균 구하는 함수
def avg_2(x, y):
    return (x + y) / 2
```

```python
print(avg_2(df['a'], df['b'])) #벡터화된 함수
```

```python
#벡터화할 수 없는 계산을 수행하는 함수
import numpy as np

def avg_2_mod(x, y):
    if (x == 20):
        return(np.nan)
    else:
        return (x + y) / 2
```

```python
# print(avg_2_mod(df['a'], df['b']))  # 오류
```

```python
#숫자 지정 -> 오류X
print(avg_2_mod(10, 20))
```

```python
#넘파이와 넘바로 벡터화하기 
#함수를 벡터 요소별로 계산하도록 수정
import numpy as np

avg_2_mod_vec = np.vectorize(avg_2_mod)
```

```python
print(avg_2_mod_vec(df['a'], df['b']))
```

```python
#데코레이터 사용: 다른 함수를 입력으로 받아 해당 함수의 출력이 동작하는 방식을 수정하는 함수
@np.vectorize
def v_avg_2_mod(x, y):
    if (x == 20):
        return(np.nan)
    else:
        return (x + y) / 2

print(v_avg_2_mod(df['a'], df['b']))
```

```python
#넘바로 벡터화하기: 배열에 적용하는 수학 계산을 최적화라도록 설계됨
import numba

@numba.vectorize
def v_avg_2_numba(x, y):
    if (int(x) == 20): #자료형 정보 추가
        return(np.nan)
    else:
        return (x + y) / 2
```

```python
#numba는 판다스 객체를 터리하는 기능 X -> 데이터프레임의 열백터 그대로 전달 시 오류 발생
# print(v_avg_2_numba(df['a'], df['b']))  # 오류
```

```python
#시리즈 객체의 values 속성 사용
print(v_avg_2_numba(df['a'].values, df['b'].values))
```
