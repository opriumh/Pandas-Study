```python
#데이터 프레임에 람다 함수 사용하기
import pandas as pd

df = pd.DataFrame({'a': [10, 20, 30],
                   'b': [20, 30, 40]})
print(df)
```

```python
def my_sq(x):
    return x ** 2

df['a_sq'] = df['a'].apply(my_sq)
print(df)
```

```python
#람다 함수: 코드 한 줄로 바로 정의할 수 있음 (apply() 함수에 람다 함수를 입력)
df['a_sq_lamb'] = df['a'].apply(lambda x: x ** 2)
print(df)
```
