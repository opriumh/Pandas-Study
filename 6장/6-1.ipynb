```python
#사용자 함수 만들기
#제곱 반환
def my_sq(x):
    return x ** 2
```

```python
#두 숫자의 평균
def avg_2(x, y):
    return (x + y) / 2
```

```python
#함수 도움말 추가
def avg_2(x, y):
    """두 숫자의 평균을 구하는 
    """ #독스트링
    return (x + y) / 2
```

```python
#함수 호출
my_calc_1 = my_sq(4)
print(my_calc_1)
```

```python
my_calc_2 = avg_2(10, 20)
print(my_calc_2)
```
