```python
def my_sq(x):
    return x ** 2
```

```python
#데이터 프레임에 함수 적용하기
import pandas as pd

df = pd.DataFrame({"a": [10, 20, 30],
                   "b": [20, 30, 40]})
print(df)
```

```python
print(df['a'] ** 2)
```

```python
#시리즈에 함수 적용하기 -> 데이터프레임에 대괄호'[]'를 덧붙이면 단일 열이나 행 일부를 추출할 수 있음=>반환 객체: 시리즈
print(type(df['a']))
```

```python
print(type(df.iloc[0]))
```

```python
sq = df['a'].apply(my_sq) #apply()메소드: 시리즈의 모든 요소에 지정한 함수 적용 가능
print(sq)
```

```python
#사용자 함수 만들어 데이터프레임에 적용하기
def my_exp(x, e):
    return x ** e
```

```python
cubed = my_exp(2, 3) #인수 두개 전달, 한개 전달 시 오류
print(cubed)
```

```python
ex = df['a'].apply(my_exp, e=2) #첫번째 매개변수: 자동 전달, 그 외: 변수와 함께 명시
print(ex)
```

```python
ex = df['a'].apply(my_exp, e=3)
print(ex)
```

```python
#데이터프레임에 함수 적용하기 -> 적용할 축을 지정(열 단위 vs 행 단위)
df = pd.DataFrame({"a": [10, 20, 30], 
                   "b": [20, 30, 40]})
print(df)
```

```python
#하나의 값을 전달받아서 출력하는 함수
def print_me(x):
    print(x)
```

```python
#열 단위로 함수 적용하기
df.apply(print_me, axis=0) #axis: 열을 대상으로 적용
```

```python
#세 숫자의 평균을 구하는 avg_3() 함수
def avg_3(x, y, z):
    return (x + y + z) / 3
```

```python
#함수 수정 (apply()의 첫 번째 인수로 전체열 전달 -> 수정 필요)
def avg_3_apply(col):
    x = col[0]
    y = col[1]
    z = col[2]
    return (x + y + z) / 3
```

```python
print(df.apply(avg_3_apply))
```

```python
#행 단위로 함수 적용
#행의 전체 값을 받아서 2개의 값을 추출하고 평균을 반환하는 avg_2_apply() 함수 적용
def avg_2_apply(row):
    x = row[0]
    y = row[1]
    return (x + y) / 2
```

```python
print(df.apply(avg_2_apply, axis=1))
```
