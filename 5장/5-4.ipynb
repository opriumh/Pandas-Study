```python
import pandas as pd

#행과 열에 있는 변수 정리하기
weather = pd.read_csv('C:/Users/minju/OneDrive/바탕 화면/Study/Coding/do_it_pandas-main/do_it_pandas-main/data/weather.csv')
print(weather.iloc[:5, :11])
```

```python
#피벗 되돌리기 (넓은 데이터가 문제가 되는 것은 아니지만 더 사용하기 쉽게 이 행위를 함)
weather_melt = weather.melt(
    id_vars=["id", "year", "month", "element"],
    var_name="day",
    value_name="temp",
)
print(weather_melt)
```

```python
#element 열에 저장된 변수를 피벗하기
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temp'
)
print(weather_tidy)
```

```python
#계층 열을 평면화
weather_tidy_flat = weather_tidy.reset_index()
print(weather_tidy_flat)
```

```python
#메서드 체인 이용 -> element 피벗 단계 간소화 가능
weather_tidy = (
    weather_melt
    .pivot_table(
        index=['id', 'year', 'month', 'day'],
        columns='element',
        values='temp')
    .reset_index()
)
print(weather_tidy)
```
