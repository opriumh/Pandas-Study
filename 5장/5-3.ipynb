```python
import pandas as pd

#열 이름이 여러 가지 뜻일 때
ebola = pd.read_csv('C:/Users/minju/OneDrive/바탕 화면/Study/Coding/do_it_pandas-main/do_it_pandas-main/data/country_timeseries.csv')
print(ebola.columns)
```

```python
#깔끔한 데이터 만들기 1
print(ebola.iloc[:5, [0, 1, 2, 10]])
```

```python
#피벗 되돌리기
ebola_long = ebola.melt(id_vars=['Date', 'Day'])
print(ebola_long)
```

```python
#열 이름 분할하고 새로운 열 할당하기
#깔끔한 데이터 만들기 2
variable_split = ebola_long.variable.str.split('_') #'_'를 기분으로 열 이름 분할
print(variable_split[:5])
```

```python
print(type(variable_split)) #세번째 요소
```

```python
print(type(variable_split[0])) #첫번째 요소
```

```python
#분할한 값을 새로운 열로 할당
status_values = variable_split.str.get(0)
country_values = variable_split.str.get(1)

print(status_values)
```

```python
#원하는 시리즈를 완성했다면 이를 데이터 프레임에 추가
ebola_long['status'] = status_values
ebola_long['country'] = country_values
print(ebola_long)
```

```python
#한 번에 분할하고 합치기
#깔끔한 데이터 한 번에 만들기
ebola_long = ebola.melt(id_vars=['Date', 'Day'])
```

```python
variable_split = ebola_long.variable.str.split('_', expand=True) #매개변수 expand에 인수 True 전달
print(variable_split)
```

```python
#판다스의 다중 할당 가능을 사용해 새로 분할한 열을 원본 데이터프레임에 바로 할당할 수 있음
ebola_long[['status', 'country']] = variable_split
print(ebola_long)
```
