```python
import pandas as pd

pew = pd.read_csv('C:/Users/minju/OneDrive/바탕 화면/Study/Coding/do_it_pandas-main/do_it_pandas-main/data/pew.csv')
print(pew)
```

```python
#넓은 데이터 확인하기
print(pew.iloc[:, 0:5]) #넓은 데이터라고 부를 것임
```

```python
#긴 데이터로 만들기
pew_long = pew.melt(id_vars='religion') #religion만 유지하고 나머지는 긴 데이터로 만듦
print(pew_long)
```

```python
pew_long = pew.melt(
    id_vars="religion", var_name="income", value_name="count"
)
print(pew_long)
```

```python
#여러 개 열 유지하기
billboard = pd.read_csv('C:/Users/minju/OneDrive/바탕 화면/Study/Coding/do_it_pandas-main/do_it_pandas-main/data/billboard.csv')
print(billboard.iloc[0:5, 0:16])
```
