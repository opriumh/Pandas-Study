```python
import seaborn as sns

tips = sns.load_dataset("tips")
```

```python
import matplotlib.pyplot as plt
```

```python
#히스토그램
fig, ax = plt.subplots()
tips['total_bill'].plot.hist(ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots()
tips[['total_bill', 'tip']].plot.hist(alpha=0.5, bins=20, ax=ax)
plt.show()
```

```python
#밀도 분포 그래프
fig, ax = plt.subplots()
tips['tip'].plot.kde(ax=ax)
plt.show()
```

```python
#산점도 그래프
fig, ax = plt.subplots()
tips.plot.scatter(x='total_bill', y='tip', ax=ax)
plt.show()
```

```python
#육각 그래프
fig, ax = plt.subplots()
tips.plot.hexbin(x='total_bill', y='tip', gridsize=10, ax=ax) #육각형 크기는 gradesize로 조정 가능
plt.show()
```

```python
#박스 그래프
fig, ax = plt.subplots()
tips.plot.box(ax=ax)
plt.show()
```
