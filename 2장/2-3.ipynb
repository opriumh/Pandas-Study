```python
#range + step 사용 (step: 간격 조정)
small_range = list(range(0, 6, 2))
subset = df.iloc[:, small_range]
print(subset)
```

```python
#range()와 :를 사용하여 열 데이터를 추출하면서 슬라이싱 구문으로 데이터를 추출하는 방법
#range
small_range = list(range(3))
subset = df.iloc[:, small_range]
print(subset)
```

```python
#슬라이싱
subset = df.iloc[:, :3]
print(subset)
```

```python
#시작값 끝값 지정
#range
small_range = list(range(3, 6))
subset = df.iloc[:, small_range]
print(subset)
```

```python
#슬라이싱
subset = df.iloc[:, 3:6]
print(subset)
```

```python
#간격을 조절하면서 추출하는 방법
#range
small_range = list(range(0, 6, 2))
subset = df.iloc[:, small_range]
print(subset)
```

```python
#슬라이싱
subset = df.iloc[:, 0:6:2] #모든 행의 0,2,4번 열을 추출
print(subset)
```

```python
#행과 열을 함께 지정하여 추출하기
print(df.loc[42, 'country'])
print(df.iloc[42, 0])
```

```python
#여러 행과 열 지정하여 데이터 추출하기
print(df.iloc[[0, 99, 999], [0, 3, 5]])
print(df.loc[[0, 99, 999], ['country', 'lifeExp', 'gdpPercap']])
```
