```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/minju/OneDrive/바탕 화면/Study/Coding/do_it_pandas-main/do_it_pandas-main/data/gapminder.tsv', sep = '\t')
```

```python
#1개 열 그룹화하기(그룹화, 열 선택, 평균 구하기)
print(df.groupby('year')['lifeExp'].mean())
```

```python
# 단계별로 그룹화보기
#그룹화
grouped_year_df = df.groupby('year')
print(type(grouped_year_df))
```

```python
print(grouped_year_df) #그래도 출력 -> 저장된 메모리 출력
```

```python
#열 선텍
grouped_year_df_lifeExp = grouped_year_df['lifeExp']
print(type(grouped_year_df_lifeExp))
```

```python
print(grouped_year_df_lifeExp)
```

```python
#평균 구하기
mean_lifeExp_by_year = grouped_year_df_lifeExp.mean()
print(mean_lifeExp_by_year)
```

```python
#2개 이상의 열 그룹화하기
multi_group_var = df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean()
```

```python
print(multi_group_var) #각 그룹이 다른 줄에 존재하게 됨 -> 앞 그룹은 계층 구조를 이룸
```

```python
#계층 -> 평탄화 (reset_index() 사용)
flat = multi_group_var.reset_index()
print(flat)
```

```python
#그룹화한 데이터 개수 세기 (nunique() 사용: 중복을 제외한 개수를 구함)
print(df.groupby('continent')['country'].nunique())
```

```python
print(df.groupby('continent')['country'].value_counts()) #value_counts(): 지정한 열이나 행의 개수를 구함
```
